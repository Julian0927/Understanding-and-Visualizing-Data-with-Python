# ***Array Math***

Basic mathematical functions operate elementwise on arrays, and are available both as operator overloads and as functions in the numpy module:

```python
x = np.array([[1,2],[3,4]], dtype=np.float64)
y = np.array([[5,6],[7,8]], dtype=np.float64)
```

```python
print(x + y)
print("\n")
print(np.add(x, y))
```

```python
print(x - y)
print("\n")
print(np.subtract(x, y))
```

```python
# This operations doesn´t matter matemhatically, but it is important in excell aplications
print(x * y)
print("\n")
print(np.multiply(x, y))
```

```python
# This operations doesn´t matter matemhatically, but it is important in excell aplications
print(x / y)
print("\n")
print(np.divide(x, y))
```

```python
# This operations doesn´t matter matemhatically, but it is important in excell aplications
print(np.sqrt(x))
```

```python
x = np.array([[1,2],[3,4]])

### Compute sum of all elements; prints "10"
print(np.sum(x))
```

```python
x = np.array([[1,2],[3,4]])


### Compute sum of each column; prints "[4 6]"
print(np.sum(x, axis=0)) 
```

```python
### Compute sum of each row; prints "[3 7]"
print(np.sum(x, axis=1))
```

```python
x = np.array([[1,2],[3,4]])

### Compute mean of all elements; prints "2.5"
print(np.mean(x))
```

```python
x = np.array([[1,2],[3,4]])

### Compute mean of each column; prints "[2 3]"
print(np.mean(x, axis=0)) 
```

```python
x = np.array([[1,2],[3,4]])

### Compute mean of each row; prints "[1.5 3.5]"
print(np.mean(x, axis=1))
```

# ***MatPlotLib***

Matplotlib is a plotting library. In this section give a brief introduction to the matplotlib.pyplot module.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Compute the x and y coordinates for points on a sine curve
pi=np.pi

def sin(x):
  return np.sin(x)

# x={0,0.1, 0.2,..., 2.9,3}
x = np.arange(0, 3 * pi, 0.1)

# y={sin(0),sin(0.1),...,sin(3)}
y = sin(x)
```

```python
# Plot the points using matplotlib
plt.plot(x, y)

# You must call plt.show() to make graphics appear.
plt.show()  
```

```python
def sin(x):
  return np.sin(x)

def cos(x):
  return np.cos(x)
```

```python
# x={0, 0.1, 0.2, ..., 2.9, 3}
x = np.arange(0, 3 * np.pi, 0.1)

# y_1={sin(0), sin(0,1),...,sin(3)}
y_1 = sin(x)

# y_2={cos(0), cos(0.1),...,cos(3)}
y_2 = cos(x)

# Plot the points using matplotlib
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

#### Subplots

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1)

# Make the first plot
plt.plot(x, y_sin)
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

# Show the figure.
plt.show()
```

# ***Scatterplots***



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Store the url string that hosts our .csv file
url = "Cartwheeldata.csv"

# Read the .csv file and store it as a pandas Data Frame
df = pd.read_csv(url)
```

```python
# Create Scatterplot
sns.lmplot(x='Wingspan', y='CWDistance', data=df)
plt.show()
```

```python
# Scatterplot arguments
sns.lmplot(x='Wingspan', y='CWDistance', data=df, fit_reg=False) 
plt.show()
```

```python
# Construct Cartwheel distance plot
sns.swarmplot(x="Gender", y="CWDistance", data=df)
plt.show()
```

# ***Boxplots***

```python
sns.boxplot(data=df.loc[:, ["Age", "Height", "Wingspan", "CWDistance", "Score"]])
plt.show()
```

```python
# Female Boxplot
sns.boxplot(data=df.loc[df['Gender'] == 'F', ["Age", "Height", "Wingspan", "CWDistance", "Score"]])
plt.show()
```
