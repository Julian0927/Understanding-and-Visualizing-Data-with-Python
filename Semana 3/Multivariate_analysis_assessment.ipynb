# ***Scatter plots***

$ \ $

First, let's looks at two variables that we expect to have a strong relationship, 'Height' and 'Wingspan'.

````python
# Make a Seaborn scatter plot (gráfico de dispersión) with x = height and y = wingspan using sns.scatterplot(x, y)
sns.lmplot(x='Height', y='Wingspan', data=df, fit_reg=False)
plt.show()
````

````python
female=df[df["Gender"]=="F"]
sns.lmplot(x='Height', y='Wingspan', data=female, fit_reg=False)
plt.show()
````

````python
male=df[df["Gender"]=="M"]
sns.lmplot(x='Height', y='Wingspan', data=male, fit_reg=False)
plt.show()
````

````python
# Create a boxplot and histogram of the tips grouped by smoking status.
plt.figure(figsize=(20,6))
sns.boxplot(data=df["CWDistance"], orient="h", color="aquamarine")
plt.show() 
````
